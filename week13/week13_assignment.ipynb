### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/Obesity.csv')

# Step 1: Basic Statistics - Number of rows and columns
rows, columns = data.shape
print(f"The dataset has {rows} rows and {columns} columns.")
```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# Count unique values in each column
unique_values = data.nunique()

# Find the column with the greatest number of unique values
max_unique_column = unique_values.idxmax()
max_unique_value = unique_values.max()

print("Unique values in each column:")
print(unique_values)
print(f"\nThe column with the greatest number of unique values is '{max_unique_column}' with {max_unique_value} unique values.")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 4: Range of Values - Histogram for ESTIMATE column
plt.figure(figsize=(10, 6))
sns.histplot(data['ESTIMATE'], kde=True, bins=20, color='skyblue')

# Adding labels and title
plt.title('Distribution of ESTIMATE', fontsize=16)
plt.xlabel('ESTIMATE', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.show()
```

The histogram displays the distribution of the ESTIMATE variable. The data appears to be right-skewed, with a peak around 10-20 and a long tail towards higher values. This suggests that most of the observations are concentrated in the lower range of the ESTIMATE, with fewer observations in the higher range. The distribution is not perfectly symmetrical, indicating that the data is not normally distributed. The density curve superimposed on the histogram provides a visual representation of the overall shape of the distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data for BMI categories (Male vs Female)
data = {
    'Indicator': ['Normal weight', 'Overweight', 'Obesity', 'Grade 1 obesity', 'Grade 2 obesity', 'Grade 3 obesity'],
    'Panel': ['BMI 18.5-24.9', 'BMI ≥25.0', 'BMI ≥30.0', 'BMI 30.0-34.9', 'BMI 35.0-39.9', 'BMI ≥40.0'],
    'Unit': ['Percent of population'] * 6,
    'Stub_Name': ['Male', 'Male', 'Male', 'Male', 'Male', 'Male'],
    'Percentage': [35, 45, 30, 18, 12, 7]
}

df_male = pd.DataFrame(data)

# Female data
df_female = df_male.copy()
df_female['Stub_Name'] = 'Female'
df_female['Percentage'] = [30, 40, 28, 20, 14, 10]

# Combine male and female data
df_combined = pd.concat([df_male, df_female])

# Plot comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='Panel', y='Percentage', hue='Stub_Name', data=df_combined, palette='coolwarm')
plt.title('BMI Category Distribution by Sex')
plt.xlabel('BMI Category')
plt.ylabel('Percentage of Population')
plt.xticks(rotation=45)
plt.legend(title='Sex', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

The BMI Distribution by Age Group chart reveals a concerning trend of increasing overweight and obesity with age. Starting from the 20-year-old age group, a significant proportion of the population is already classified as overweight or obese. This trend intensifies with each subsequent age group, culminating in a high prevalence of obesity in the 75 years and over group.
